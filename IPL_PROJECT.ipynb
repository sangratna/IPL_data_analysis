```python
df_toss = df2[['Team1', 'Team2','TossWinner','WinningTeam']]
```

```python
df_toss['TossWinMatchWin'] = df_toss['TossWinner'] == df_toss['WinningTeam']
df_toss
```

```python
result_counts = df_toss['TossWinMatchWin'].value_counts()
result_counts
```

```python
plt.figure(figsize=(6, 6))
plt.pie(result_counts, labels=['Toss Win & Match Win', 'Toss Win & Match Loss'], autopct='%1.1f%%', startangle=140, colors=['green', 'red'])
plt.title('Toss Winning Team Match Outcome')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
matches_played=df2['Team1'].value_counts()+df2['Team2'].value_counts()
matches_played
```

```python
matches_played_df = matches_played.to_frame().reset_index()
matches_played_df.columns=['Team_names','Matches_played']
```

```python
matches_played_df
```

```python
wins= pd.DataFrame(df2['WinningTeam'].value_counts()).reset_index()
```

```python
wins.columns=['Team_names','wins']
wins
```

```python
played=matches_played_df.merge(wins,on='Team_names',how='inner')
```

```python
played
```

```python
import plotly.graph_objs as go
trace1 = go.Bar(x=played['Team_names'], y=played['Matches_played'],name='Total_matches')
trace2 = go.Bar(x=played['Team_names'], y=played['wins'],name='Matches_won')
```

```python
data=[trace1, trace2]
```
